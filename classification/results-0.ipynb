```python
import pandas as pd
history_autoencoder = pd.read_csv('gdrive/My Drive/UPJV/missing/history_ae-0.csv')
history_mlp = pd.read_csv('gdrive/My Drive/UPJV/missing/history_mlp-0.csv')
history_gnn = pd.read_csv('gdrive/My Drive/UPJV/missing/history_gnn-4.csv')
```

```python
import matplotlib.pyplot as plt

# Plot training & validation loss values
plt.figure(figsize=(12, 6))

plt.plot(history_gnn['loss'], label='GNN Train Loss', linewidth='4', linestyle="--")
plt.plot(history_gnn['val_loss'], label='GNN Validation Loss', linewidth='4', linestyle=":")

plt.plot(history_autoencoder['loss'], label='Autoencoder Train Loss', linewidth='4', linestyle="--")
plt.plot(history_autoencoder['val_loss'], label='Autoencoder Validation Loss', linewidth='4', linestyle=":")

plt.plot(history_mlp['loss'], label='MLP Train Loss', linewidth='4', linestyle="--")
plt.plot(history_mlp['val_loss'], label='MLP Validation Loss', linewidth='4', linestyle=":")

plt.rc('font', size=22)
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(loc='upper right')
plt.show()
```
